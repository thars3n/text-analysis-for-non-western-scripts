```python
import nltk
```

```python
# directory containing all source texts
data_dir="../corpus"
```

```python
import os, glob
os.chdir(data_dir)
docnames = []
documents = []
charcount = 0
bytescount = 0
alldocs = ""

for filename in glob.glob("*.txt"):
    filedata = open(filename, 'r').read()
    docnames.append(filename)
    bytescount = bytescount + len(filedata.encode('utf-8'))
    print(filename + " = " + str(len(filedata)) + " chars & bytes = " + str(len(filedata.encode('utf-8'))))
    documents.append(filedata)
    charcount = charcount + len(filedata)
    alldocs = alldocs + " " + filedata
```

```python
print("bytes " + str(bytescount))
print("\n")
print("chars " + str(charcount))
print("\n")
#print(documents[0])
```

```python
# create Text object from corpus
from nltk.text import Text
nltk_tokens = nltk.word_tokenize(alldocs)
textList = Text(nltk_tokens)
textList.concordance("methinks")
```

```python
import matplotlib
textList.dispersion_plot(["methinks","forsooth"])
```

```python
# lexical diversity scores
ld_tuple = []

def ldiv(text):
    return len(set(text)) / len(text)

xcount = 1
for doc in documents:
    nltk_tokenz = nltk.word_tokenize(doc)
    textListz = Text(nltk_tokenz)
    ld = ldiv(textListz)
    ld_tuple.append([docnames[xcount-1], ld])
    xcount = xcount + 1
    
ld_tuple

#display tuple sorted by lexical diversity score
#sorted(ld_tuple, key=lambda x: x[1], reverse=True)
```

```python
# collocations (2 words in sequence occurring more than once) from all documents
xcount = 1
for doc in documents:
    nltk_tokenz = nltk.word_tokenize(doc)
    textListz = Text(nltk_tokenz)
    #textListz.collocations()
    print("text" + str(xcount))
    print(textListz.collocations())
    print("----")
    xcount = xcount + 1
```

```python
# word lengths > 18 from all documents
wl18 = []
for doc in documents:
    nltk_tokenz = nltk.word_tokenize(doc)
    for token in nltk_tokenz:
        if len(token) > 18:
            wl18.append(token + " " + str(len(token))) 
print(wl18)
```

```python
# Use RegEx to find all words beginning with "c" and ending with "le" or "um"
import re
reoutput = []
for doc in documents:
    nltk_tokenz = nltk.word_tokenize(doc)
    for token in nltk_tokenz:
        if re.search("^c.*(le|um)$", token) and token not in reoutput:
            reoutput.append(token) 
print(reoutput)
```
